```python
import tensorflow as tf
from PIL import Image
from keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
print(tf.__version__)
```

```python
model_name ='Fold1_train/best_model.h5'

image_path = 'ETT_v3/Fold1/test/A_0_5.jpg'
mask_path= 'ETT_v3/Fold1/testannot/A_0_5.png'

img = image.load_img(image_path, target_size=(256, 256),color_mode='grayscale')
mask= image.load_img(mask_path, target_size=(256, 256),color_mode='grayscale')
img_array = image.img_to_array(img)
img_array = np.expand_dims(img_array, axis=0) 
img_array = img_array / 255.0
```

```python
model = load_model(model_name)
prediction=model.predict(img_array,verbose=1)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# 在第一个子图中绘制原始图像
axs[0].imshow(img)
axs[0].set_title('Original Image')
axs[0].axis('off')

# 在第二个子图中绘制地面真实掩码
axs[1].imshow(mask, cmap='gray')
axs[1].set_title('Ground Truth Mask')
axs[1].axis('off')

# 在第三个子图中绘制预测掩码
axs[2].imshow(prediction[0, :, :, 0], cmap='gray')
axs[2].set_title('Predicted Mask')
axs[2].axis('off')

# 显示图像
plt.show()
```
